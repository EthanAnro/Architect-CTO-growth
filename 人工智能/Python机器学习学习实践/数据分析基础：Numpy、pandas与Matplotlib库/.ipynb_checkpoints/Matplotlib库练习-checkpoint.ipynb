**NumPy库和pandas库主要用于处理一维及二维的表格数据，而Matplotlib库是数据可视化的工具。**

```python
import matplotlib.pyplot as plt
x = [1, 2, 3]
y = [2, 4, 6]
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([1, 2, 3])
y1 = x1 + 1
plt.plot(x1, y1, label='y=x+1')

y2 = x1*2
plt.plot(x1, y2, color = 'red', linewidth=3, linestyle='--', label='y=x*2')
plt.legend(loc='upper left')
plt.show()
```

```python
# (2) 柱状图
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
plt.bar(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(10)
y = np.random.rand(10)
plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np  
data = np.random.randn(10000)
plt.hist(data, bins=40, edgecolor='black')
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(data)
df.hist(bins=40, edgecolor='black')
```

```python
import pandas as pd
df = pd.DataFrame([[8000, 6000], [7000, 5000], [6500, 4000]], columns=['人均收入', '人均支出'], index=['北京', '上海', '广州'])
print(df)

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
df['人均收入'].plot(kind='pie')
df['人均收入'].plot(kind='box')
```

```python
import matplotlib.pyplot as plt

x= [1,2,3]
y= [2,4,6]

plt.plot(x, y)

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title('title')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
import numpy as np
import matplotlib.pylab as plt

x1 = np.array([10, 20, 30])
y1 = x1 
plt.plot(x1, y1, color='red', linestyle='--', label='y=x')
plt.legend(loc='upper left')

plt.twinx()

y2 = x1*x1
plt.plot(x1, y2, label='y=x^2')
plt.legend(loc='upper right')

plt.rcParams['figure.figsize'] = (8, 6)
pl.xticks(rotation=20)

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimSun']
plt.rcParams['axes.unicode_minus'] = False
x = [1, 2, 3]
y = [2, 4, 6]
plt.plot(x, y)
plt.title('中文标题')
plt.xlabel('中文X轴')
plt.ylabel('中文Y轴')
plt.show()
```

```python
# (7) 绘制多图
import matplotlib.pyplot as plt
# 绘制第一个子图：折线图
ax1 = plt.subplot(221)  
plt.plot([1, 2, 3], [2, 4, 6])
# 绘制第二个子图：柱状图
ax2 = plt.subplot(222)  
plt.bar([1, 2, 3], [2, 4, 6])
# 绘制第三个子图：散点图
ax3 = plt.subplot(223)  
plt.scatter([1, 3, 5], [2, 4, 6])
# 绘制第四个子图：直方图
ax4 = plt.subplot(224)  
plt.hist([2, 2, 2, 3, 4])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = axes.flatten()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
ax1, ax2, ax3, ax4 = axes.flatten()
ax1.plot([1, 2, 3], [2, 4, 6])
ax2.bar([1, 2, 3], [2, 4, 6])
ax3.scatter([1, 3, 5], [2, 4, 6])
ax4.hist([2, 2, 2, 3, 4])
```
